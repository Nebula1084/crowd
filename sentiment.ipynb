```python
import pandas as pd
import nlp.sanitize as st
import nlp.embedding as eb

tweet = pd.ExcelFile('data/tweet.xlsx')
data = tweet.parse('Data')
data = st.standardize(data, 'tweet_text')
text = data['tweet_text']
print(text[3])
len(text)
```

```python
from textblob import TextBlob
import numpy as np
score=np.zeros(len(text))
for i in range(len(text)):
    blob = TextBlob(text[i])
    score[i]=blob.sentiment.polarity
    print(score[i])
```

```python
writer = pd.ExcelWriter('./data/sentiment.xlsx')
df1 = pd.Series(score)
df1.to_excel(writer,'Sheet1')
writer.save()
```
